```python
import numpy as np
import matplotlib.pyplot as plt
```

# Números complexos

```python
z = 1 + 1j
print(f'Parte real: {np.real(z)}')
print(f'Parte imaginária: {np.imag(z)}')
```

```python
z = 1 - 100j
print(f'Parte real: {np.real(z)}')
print(f'Parte imaginária: {np.imag(z)}')
```

```python
z2 = 1 + 10j
```

```python
plt.plot(z.real, z.imag, 'ro')
plt.plot(z2.real, z2.imag, 'ks')
plt.plot((z+z2).real, (z+z2).imag, 'pb')
plt.xlabel("componente real")
plt.ylabel("componente imaginária")
plt.title("Gerar um número complexo no plano")
plt.show()
```

```python
t = np.linspace(0, 2*np.pi)
plt.plot(t, np.cos(t), '-r')
plt.plot(t, np.sin(t), '-b')
plt.xlabel('t')
plt.title("Funções trigonométricas")
plt.show()
```

# Questão 1

```python
t = np.linspace(0, 10, 200)
z = -1
plt.plot(t, np.exp(z*t), 'r-')
plt.title('"plots" envolvendo a função exponencial')
z = 1e-1
plt.plot(t, np.exp(z*t), 'k--')
z = 5j
plt.plot(t, np.exp(z*t), 'b-.')
z = -1 + 5j
plt.plot(t, np.exp(z*t), 'k:')
plt.xlabel('t')
plt.show()
```

```python
plt.plot(t, np.real(np.exp(z*t)), 'b-', label='real')
plt.plot(t, np.imag(np.exp(z*t)), 'g-', label='img')
plt.xlabel('t')
plt.title('"plots" parte real e parte imaginária da função exponencial')
plt.legend()
plt.show()
```

# Questão 2

```python
t = np.linspace(0, 10, 200)
z = 1j
plt.plot(np.exp(z*t).real, np.exp(z*t).imag, 'r-')
plt.title('"plots" envolvendo a função exponencial')
z = -.1 + 1j
plt.plot(np.exp(z*t).real, np.exp(z*t).imag, 'g--')
z = .1 + 1j
plt.plot(np.exp(z*t).real, np.exp(z*t).imag, 'b-.')
plt.xlabel('parte real')
plt.ylabel('parte imaginária')
plt.show()
```
