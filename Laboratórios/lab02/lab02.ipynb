```python
import numpy as np
import matplotlib.pyplot as plt
import sounddevice as sd
```

## Exercicio 1

```python
fs = 44100
t = np.arange(0, 2, 1/fs)
f = 1000
A = 0.2
w = 15 * np.pi/180
y = A * np.sin(2 * np.pi * f * t + w)
plt.plot(t, y)
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude')
plt.show()
sd.play(y, fs)
sd.wait()
```

## Exercicio 2

```python
# Cancelamento de fase
fs = 44100
t = np.arange(0, 5, 1/fs)
f1, f2 = 440, 441
A1, A2 = 0.3, 0.3
w1, w2 = 0, 180
y1 = A1 * np.sin(2 * np.pi * f1 * t + np.deg2rad(w1))
y2 = A2 * np.sin(2 * np.pi * f2 * t + np.deg2rad(w2))
y = (y1 + y2) / 2
plt.plot(t,y)
sd.play(y, fs)
sd.wait()
```

## Exercicio 3

```python
# Efeito de batidas
fs = 44100
t = np.arange(0, 5, 1/fs)
f1, f2 = 300, 310
# f1, f2 = 300, 302
# f1, f2 = 1000, 1010
# f1, f2 = 300, 330
A = 0.5
w = 0
y1 = A * np.sin(2 * np.pi * f1 * t + np.deg2rad(w))
y2 = A * np.sin(2 * np.pi * f2 * t + np.deg2rad(w))
y = np.vstack((y1, y2))
sd.play(y.T,fs)
sd.wait()
```

## Exercicio 4

```python
fs = 44100
t = np.arange(0, 5, 1/fs)

frequencies = [130.813, 195.998, 261.626, 329.628, 391.995]
amplitudes = [0.3, 0.3/2, 0.3/3, 0.3/4, 0.3/5]
y = np.zeros(len(t))

for i in range(len(frequencies)):
    y += amplitudes[i] * np.sin(2 * np.pi * frequencies[i] * t)

y /= len(frequencies)  # Normalizando a amplitude
plt.plot(t,y)
sd.play(y, fs)
sd.wait()
```
